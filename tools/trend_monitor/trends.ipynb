```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from trendspy import Trends


def extract_visible_text(url):
    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove all script and style elements
        for script_or_style in soup(['script', 'style']):
            script_or_style.decompose()

        # Get all visible text
        visible_text = soup.get_text(separator=' ', strip=True)
        print('Suceess ! ')
        return visible_text
    else:
        print(f"Failed to fetch the webpage. Status code: {response.status_code}")
        print(response.__dict__)
        return None

# Example usage
url = "https://www.versus.com.py/hoy/2025/02/12/afirman-que-carlos-paul-tiene-algo-contra-luqueno-no-es-la-primera-vez/"
text = extract_visible_text(url)

if text:
    print("Extracted Text:\n")
    print(text)

tr = Trends()
```

```python
trends_last_48 = tr.trending_now(geo='MX', hours=48)
```

```python
data_list = []
for trend in trends_last_48:
    data_list.append(trend.__dict__)
df_trends_48 = pd.DataFrame(data_list)
```

```python
df_trends_48.sort_values('volume', ascending=False, inplace=True)
df_trends_48 = df_trends_48.head(50).reset_index(drop=True)
```

```python
news_dict = {}

for index, row in df_trends_48.iterrows():
    print('finding for ', row['keyword'])
    news = tr.trending_now_news_by_ids(row['news_tokens'], max_news=3)
    print(f' --> Getting news for {row["keyword"]}')
    news_dict[index] = [news.__dict__ for news in news]  # Store news for each trend with its index

# Add the news column to the DataFrame
df_trends_48['news'] = pd.Series(news_dict)
```

```python
for idx, news in enumerate(df_trends_48['news']):
    print(f'\nProcessing trend {idx}:')
    for n in news:
        print(f'[{idx}] extracting text from:', n['url'])
        n['text'] = extract_visible_text(n['url'])
```
